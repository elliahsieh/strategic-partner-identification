```python
# Convert the mapping dataframe into a dictionary for faster lookup

mapping_dict = dict(zip(institution_type_mapping['Institution'], institution_type_mapping['Category']))

# Step 1: Create the 'Institutions' column by combining Sponsor and Collaborators

df['Institutions'] = df['Sponsor'].fillna('') + '|' + df['Collaborators'].fillna('')

# Step 2: Define the function to map institutions to categories and remove duplicates

def map_institution_category(row, mapping_dict):
    # Split by '|' and map each institution to its category
    institution_list = row.split('|')
    category_list = [mapping_dict.get(inst.strip(), 'Unknown') for inst in institution_list if inst.strip()]
    
    # Remove duplicates and keep the order
    unique_categories = list(dict.fromkeys(category_list))  # Removes duplicates while preserving order
    
    # Join the unique categories by ', '
    return ', '.join(unique_categories)

# Apply the function to the 'Institutions' column

df['Institution Categories'] = df['Institutions'].apply(map_institution_category, mapping_dict=mapping_dict)

# Step 3: Drop the 'Institutions' column after categorization

df.drop(columns=['Institutions'], inplace=True)

# Display the final dataframe

print(df)
```
